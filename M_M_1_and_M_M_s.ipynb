# **M/M/1**

Input parameters

```python
'''
input parameter :
- lambda (la)   : arrival rate
- mu (mu)       : service rate, if known
- n             : for calculating probability of n customers
- m             : m-th percentile
'''

la = 4
#mu = 75          #uncomment if mu is known
n = 11
m = 90
```

If mu is unknown, and average service time is known, then mu is:

```python
# average service time (1/mu)
# ''' uncoment this if mu is known

ast = 0.22

mu = 1/ast
print(f'mu = {mu}')

# uncomment this if mu is known ''' 
```

Utilization factor (U or rho) = percentage of the time that all servers are busy or known as **queueing intensity**



image.png

```python
rho = la/mu
print(f'rho = {rho}')
```

P0 = probability that there are no customers in the system

image.png

```python
p0 = 1 - rho
print(f'p0 = {p0*100} %')
```

Wq = average time a customer spends in waiting line waiting for service or known as **delay in queue**

image.png

```python
wq = la / (mu*(mu - la))
print(f'Wq = {wq}')
```

W = average time a customer spends in the system (in waiting line and being service) or known as **delay in system**

image.png

```python
w = 1 / (mu - la)
print(f'W = {w}')
```

Lq = average number of customer in waiting line for service or known as **queue length in queue**

image.png

```python
lq = (la**2) / (mu * (mu - la))
print(f'Lq = {lq}')
```

L = average number of customer in the system (in waiting line and being served) or known as **queue length in system**

image.png

```python
l = rho / (1 - rho)
print(f'L = {l}')
```

Pn = probability that there are n customers in the system

image.png

```python
pn = (rho**n) * p0
print(f'rho^{n} = {rho**n}')
print(f'P{n} = {pn}')
```

phi_q(m) = m-th percentile waiting time in queue

```python
phi_q = w * math.log((100 - m) * rho)
print(f'phi_q({m}) = {phi_q}')
```

# **M/M/s**

```python
'''
input parameter :
- lambda (la)   : arrival rate
- mu (mu)       : service rate, if known
- s             : number of servers
- n             : for calculating probability of n customers
'''

la = 4
#mu = 75          #uncomment if mu is known
s = 1
n = 11
```

If mu is unknown, and average service time is known, then mu is:

```python
# average service time (1/mu)
#''' uncoment this if mu is known

ast = 0.22

mu = 1/ast
print(f'mu = {mu}')

#uncomment this if mu is known ''' 
```

Utilization factor (U or rho) = percentage of the time that all servers are busy or known as queueing intensity

image.png

```python
rho = la / (s*mu)
print(f'rho = {rho}')
```

P0 = probability that there are no customers in the system

image.png

```python
i = 0
left = 0

while i <= s-1:
  forLeft = ((la/mu)**i) / (math.factorial(int(i)))
  left += forLeft
  i+=1

center = ((la/mu)**s) / (math.factorial(int(s)))
right = 1 / (1 - (rho))

p0 = 1 / (left + center * right)
print(f'P0 = {p0*100} %')
```

Lq = average number of customer in waiting line for service or known as **queue length in queue**

image.png

```python
lq = p0 * (((la/mu)**s) / math.factorial(int(s))) * ((rho) / ((1-rho)**2))
print(f'Lq = {lq}')
```

Wq = average time a customer spends in waiting line waiting for service or known as **delay in queue**

image.png

```python
wq = lq/la
print(f'Wq = {wq}')
```

W = average time a customer spends in the system (in waiting line and being service) or known as **delay in system**

image.png

```python
w = wq + 1/mu
print(f'W = {w}')
```

L = average number of customer in the system (in waiting line and being served) or known as **queue length in system**

image.png

```python
l = lq + la/mu
print(f'L = {l}')
```

Pn = probability that there are n customers in the system

image.png

```python
if s <= n:
  cn = ((la/mu)**n) / ((math.factorial(int(s)) * s**(n-s)))
else:
  cn = ((la/mu)**n) / (math.factorial(int(n)))

pn = cn * p0

print(f'C{n} = {cn}')
print(f'P{n} = {pn}')
```
